### Análise de Dados: Síndrome dos Ovários Policísticos (PCOS)

## Etapa 1: Importação e leitura dos dados

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("pcos_rotterdam_balanceado.csv")
df.head()
df.info()
df.describe()
df.shape
df.columns
```

```python
df.columns = ['age', 'bmi', 'menstrual_irregularity', 'testosterone', 'follicle_count', 'pcos_diagnosis']
df['menstrual_irregularity'].value_counts()
df['pcos_diagnosis'].value_counts()
sns.boxplot(data=df[['bmi', 'testosterone', 'follicle_count']])
```

```python
# Relação entre testosterona e diagnóstico
sns.boxplot(x='pcos_diagnosis', y='testosterone', data=df)
```

```python
# Relação entre BMI e diagnóstico
sns.violinplot(x='pcos_diagnosis', y='bmi', data=df)
```

```python
# Correlação geral
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
df.to_csv("pcos_dados_tratados.csv", index=False)
```

## Conclusão

A análise revelou que há uma associação entre altos níveis de testosterona e diagnóstico de PCOS. Além disso, observou-se que pacientes com PCOS tendem a apresentar IMC e contagem de folículos antrais mais elevados.

Este estudo pode auxiliar na compreensão de fatores importantes para o diagnóstico precoce da síndrome.
